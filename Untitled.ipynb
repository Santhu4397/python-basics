```python
## LinkedList
 
class Node():

    def __init__(self, data):

        self.data = data

        self.next = None

        self.prev = None
 
class LinkedList():

    def __init__(self):

        self.head = None

        self.tail = None
 
    def insert(self, data):

        new_node = Node(data)
 
        if(self.head == None):

            self.head = new_node

            self.tail = new_node

            return
 
        element = self.head

        while (element.next != None):

            element = element.next
 
        element.next = new_node

        new_node.prev = element
 
    
 
```

```python
import collections 

class Bike:
    def __init__(self, make, model, features, cost, color):
        self.make = make
        self.model = model
        self.features = features
        self.cost = cost
        self.color = color

    def __str__(self):
        return f"Make: {self.make}, Model : {self.model}, Cost : Rs. {self.cost} /-, Color : {self.color}, \nFeature: {self.features}"

    
    # Merge sort for bikes based on cost
    @staticmethod
    def mergeSort(bikes):
        if len(bikes) <= 1:
            return bikes
    
        mid = len(bikes) // 2
        leftHalf = bikes[:mid]
        rightHalf = bikes[mid:]
        sortedLeft = Bike.mergeSort(leftHalf)
        sortedRight = Bike.mergeSort(rightHalf)
        return Bike.merge(sortedLeft, sortedRight)
    
    def merge(left, right):
        result = []
        i = j = 0
       
        while i < len(left) and j < len(right):
            if left[i].cost < right[j].cost:
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1
    
        result.extend(left[i:])
        result.extend(right[j:])
    
        return result

class Feature:
    def __init__(self, milage, brake, cc, type):
        self.milage = milage
        self.brake = brake
        self.cc = cc
        self.type = type

# Define bikes
hondaPlatinafeature = Feature(80, 'disc', '150CC', 'gear')
hondabike1 = Bike("honda", "x2", hondaPlatinafeature, 75_000, 'Black')
hondabike2 = Bike("honda", "x2", hondaPlatinafeature, 70_000, 'Blue')
yamahaFeature = Feature(50, 'disc', '200CC', 'gear')
yamaha = Bike("Yamaha", "RX100", yamahaFeature, 1_50_000, 'Black')
yamahaScootyFeature = Feature(60, 'normal', '100CC', 'moppet')
scooty = Bike("Yamaha", "rayzr 125", yamahaScootyFeature, 90_000, 'Red')
tvsFeature = Feature(70, 'normal', '120CC', 'gear')
tvs = Bike("TVS", "Star city", tvsFeature, 50_000, 'Red&Blue')

# Add bikes to showroom
bikes = collections.deque([hondabike1, hondabike2, yamaha, scooty, tvs])


# Sort and print bikes by cost
sortedBikes =Bike.mergeSort(list(bikes))  
print("size",len(sortedBikes))
for bike in sortedBikes:
    print(bike)
```
